```python
from bs4 import BeautifulSoup
```

```python
import requests

url = 'http://www.imdb.com/chart/bottom'

response = requests.get(url)
```

```python
response
```

```python
page = response.text
soup = BeautifulSoup(page, "lxml")
```

```python
print (soup.prettify())
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
soup.select('.titleColumn')[0].a.text #movie title, range(100)
```

```python
titles = []
for i in range(100):
    titles.append(soup.select('.titleColumn')[i].a.text)
    
```

```python
print(titles)
```

```python
#movie link
urls = []
for j in range(100):
    urls.append("http://www.imdb.com/title/" + str(soup.select('.titleColumn')[j].a).split('/')[2])
```

```python
#movie year
years = []
for k in range(100):
    years.append(int(str(soup.select('.secondaryInfo')[k].text)[1:5]))
```

```python
worst_movies = zip(titles, years, urls)
```

```python
new_list = []

for i in zip(titles,years,urls):
    new_list.append(i)
```
